        = probability that no item is defective   
        + probability that exactly one item is defective

        = $0.5987369392383787 + 0.03151247048623045 + (1-\frac{19}{20})^2(\frac{19}{20})^8$ = $0.6319079608028317$ ~ $0.632$
        
Q2.d: Probability that three or more of the items are defective = 1-Probability that two or fewer items are defective = $1-0.6319079608028317$ = $0.36809203919716826$ ~ $0.368$

1. <u>Win-LossFactor</u>: It is the difference of the wins and the losses by a team divided by the total games played by the team. <br>

2. <u>ScoreFactor</u>: It is the ratio of baskets scored to baskets conceded.<br>
top teams: 2, 1, 3, 4, 5. <br>
worst teams: 57, 50, 48, 60, 55. 
3. <u>AvgBasketFactor</u>: It is the difference between baskets scored and baskets given divided by the total games played.<br> 
top teams: 1, 2, 3, 4, 5.<br>
worst teams : 47, 50, 60, 59, 55.
4. <u>TournamentPerformance</u>: It is the weighted average of the number of times the team has been in the top 2 positions. Only 15 teams have a nonzero score in this metric.the top teams are 1, 2, 3, 5 and 4. 
5. <u>Oldest team</u>: The oldest teams are ones from 1929 while the newest team is as recent as 2017-18. 
6. <u>HighestPositionHeld</u>: Teams 1, 2, 3, 4, 5, 6, 8, 10 have all held the first position at least once and teams 21, 20, 14, 15 and 9 have held a runner up position at their best. 
7. <u>AvgScore</u>: It is calculated as Score divided by PlayedGames.<br> 
top teams: 1, 2, 21, 3, 4.<br>
worst teams: 58, 50, 59, 55, 60.

```python
mydata["Score"] = mydata["Score"].astype('int64')
mydata["PlayedGames"] = mydata["PlayedGames"].astype('int64')
mydata["Tournament"] = mydata["Tournament"].astype('int64')
mydata["PlayedGames"] = mydata["PlayedGames"].astype('int64')
mydata["WonGames"] = mydata["WonGames"].astype('int64')
mydata["DrawnGames"] = mydata["DrawnGames"].astype('int64')
mydata["LostGames"] = mydata["LostGames"].astype('int64')
mydata["BasketScored"] = mydata["BasketScored"].astype('int64')
mydata["BasketGiven"] = mydata["BasketGiven"].astype('int64')
mydata["TournamentChampion"] = mydata["TournamentChampion"].astype('int64')
mydata["Runner-up"] = mydata["Runner-up"].astype('int64')
```

```python
mydata.dtypes
```

```python
mydata["Win-LossFactor"] = (mydata["WonGames"]-mydata["LostGames"])/mydata["PlayedGames"]
```

```python
mydata["Win-LossFactor"].sort_values(ascending = False) 
# team number = 1 + index number  
```
